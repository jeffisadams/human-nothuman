```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
data = pd.read_csv('./culledBottles.csv')

data.head()
```

```python
row = data.iloc[0]
img = cv.imread(row.path, 0)

mask = np.zeros(img.shape, dtype=np.int16)
mask[row.ymin:row.ymax, row.xmin:row.xmax] = 1

plt.imshow(img * mask, cmap='gray')
```
